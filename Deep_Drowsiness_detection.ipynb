# Detections using custom data:



# 1.Create images using webcam:

```python
labels=['awake','drowsy']

import os
import cv2
import time

cap=cv2.VideoCapture(0)
for label in labels:
    label_dir=os.path.join('data','images',label)
    print(label_dir)
    
    time.sleep(10)
    for i in range(25):
        ret,frame=cap.read()
        cv2.imshow(label,frame)
        if not os.path.exists(label_dir):
            os.makedirs(label_dir)
        image_name=label_dir+'/'+label+'_'+str(i)+'.jpg'
        time.sleep(2)
        cv2.imwrite(image_name,frame)
        print("writing images in dir ", image_name)
        
        if cv2.waitKey(10) & 0xFF ==ord('q'):
            break        
            
cap.release()
cv2.destroyAllWindows()
```

```python
cap = cv2.VideoCapture(0)
while cap.isOpened():
    ret, frame = cap.read()
    
    # Make detections 
    video_results = custom_model(frame)
    
    cv2.imshow('YOLO', np.squeeze(video_results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break
cap.release()
cv2.destroyAllWindows()
```
